```python
import easyocr
import os
import csv

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'])

# Folder containing images
image_folder = '/content/drive/MyDrive/Colab Notebooks/train_2_images'

# CSV output file
csv_output = '/content/drive/MyDrive/Colab Notebooks/ocr_texts.csv'

# Function to extract text from an image using OCR
def extract_text_from_image(image_path):
    # Extract text using EasyOCR
    result = reader.readtext(image_path, detail=0)

# Join the list of text lines into a single string
    extracted_text = ' '.join(result)
    return extracted_text

# Function to process all images in a folder and save the results to CSV
def process_images_and_save_to_csv(image_folder, csv_output):
    # Open CSV file for writing
    with open(csv_output, mode='w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        # Write CSV header
        csv_writer.writerow(['Image Name', 'OCR Text'])

        # Loop through all files in the image folder
        for image_file in os.listdir(image_folder):
            image_path = os.path.join(image_folder, image_file)
            # Check if the file is an image (assuming common image extensions)
            if image_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                print(f"Processing image: {image_file}")
                # Extract text from the image using OCR
                ocr_text = extract_text_from_image(image_path)
                # Write the results to the CSV file
                csv_writer.writerow([image_file, ocr_text])

# Run the processing function
process_images_and_save_to_csv(image_folder, csv_output)

print(f"OCR texts saved to {csv_output}")
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/merged_data.csv')
```

```python
import re

def find_entity_indices(text, entity_value):
    # Use regex to find the start and end indices
    match = re.search(re.escape(entity_value), text)
    if match:
        start = match.start()
        end = match.end()
        return start, end
    return -1, -1

# Prepare the training data
training_data = []

for index, row in df.iterrows():
  # Extract and ensure fields are strings
    ocr_text = str(row['OCR Text']) if pd.notna(row['OCR Text']) else ""
    entity_name = str(row['entity_name']) if pd.notna(row['entity_name']) else ""
    entity_value = str(row['entity_value']) if pd.notna(row['entity_value']) else ""

    # Find start and end indices
    start, end = find_entity_indices(ocr_text, entity_value)

    # Ensure indices are valid
    if start != -1 and end != -1:
        training_data.append({
            'text': ocr_text,
            'entities': [(start, end, entity_name.upper())]
        })

# Print a sample to verify
for i in range(min(10, len(training_data))):
    print(training_data[i])
```

```python
from spacy.tokens import DocBin
from tqdm import tqdm

nlp = spacy.blank("en") # load a new spacy model
doc_bin = DocBin()
```

```python
from spacy.util import filter_spans
from tqdm import tqdm

for training_example  in tqdm(training_data):
    text = training_example['text']
    labels = training_example['entities']
    doc = nlp.make_doc(text)
    ents = []
    for start, end, label in labels:
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents
    doc_bin.add(doc)

# Save the DocBin
doc_bin.to_disk("/content/drive/MyDrive/Colab_Notebooks/train.spacy")
```
